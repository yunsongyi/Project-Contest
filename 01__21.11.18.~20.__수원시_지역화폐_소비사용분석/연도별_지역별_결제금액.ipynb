```python
Gwonseon_jobs_food=Gwonseon_jobs.loc['일반·휴게음식']
Gwonseon_jobs_food
```

```python
Gwonseon_jobs_food_gender=Gwonseon_jobs_food.set_index('성별')
Gwonseon_jobs_food_gender
```

```python
Gwonseon_jobs_food_men= Gwonseon_jobs_food_gender.loc['남']
Gwonseon_jobs_food_men
```

```python
Gmen_sum=Gwonseon_jobs_food_men['결제건수'].sum()
Gmen_sum
```

```python
Gwonseon_jobs_food_women= Gwonseon_jobs_food_gender.loc['여']
Gwonseon_jobs_food_women
```

```python
Gwomen_sum=Gwonseon_jobs_food_women['결제건수'].sum()
Gwomen_sum
```

```python
Ggender=pd.DataFrame([['남',Gmen_sum],['여',Gwomen_sum]],columns=['gender','count'])
Ggender
```

```python
Ggender_dummies=pd.get_dummies(Ggender['gender'])

Ggender_new = pd.concat([Ggender,Ggender_dummies],axis=1)
del Ggender_new['gender']
```

```python
Ggender_new=pd.DataFrame(Ggender_new)
Ggender_new
```
