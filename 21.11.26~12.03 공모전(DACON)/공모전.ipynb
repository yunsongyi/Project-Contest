# 년도별 가정폭력-아동학대 건수 그래프

```python
total = pd.read_csv('total_data.csv')
total

a = total[['년도']]
x = total[['가정폭력건수']]
y = total[['아동학대건수']]
```

```python
from matplotlib import font_manager, rc
import pandas as pd

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgunbd.ttf'
font_name = fm.FontProperties(fname=font_location).get_name()
plt.rc('font', family=font_name)


plt.rcParams["figure.figsize"] = (6,6)

ax=total.plot(kind='bar', x='년도', y='가정폭력건수', color='#FFE146')
ax2=total.plot(kind='line', 
               linewidth = 5,x='년도', y='아동학대건수', secondary_y=True,color='#FF7F50',ax = ax)

ax.set_ylabel('가정폭력건수', rotation = 0, size = 15)
ax2.set_ylabel('아동학대건수', rotation = 0, size = 15)

ax.legend(loc = 'upper left')
ax2.legend(["아동학대건수"],loc = 'upper right')

plt.title('< 연도별 가정폭력-아동 폭력 건수 >', fontsize = 22)
plt.grid(linewidth = 0.1, color = 'black')
plt.show()
```
