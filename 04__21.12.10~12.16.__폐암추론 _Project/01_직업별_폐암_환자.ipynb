# 직업별 폐암 발병 

```python
import pandas as pd
df = pd.read_csv('lung_job.csv')
```

```python
df['statsTrgtNm'] = df['statsTrgtNm'].fillna('기타') 
```

```python
df.isnull().any()
```

```python
df
```

### 직업별 그룹화

```python
df_group = df.groupby('statsTrgtNm')
print(df_group)
```

```python
df_g_count = df_group.count()
```

```python
df_g_count.to_csv('job_group.csv')
```

```python
df_new = pd.read_csv('job_group.csv')
df_new
```

```python
df_new_j = df_new[['statsTrgtNm', 'centerNm']]
df_new_j
```

### 직업별 폐암 환자

```python
import matplotlib.pyplot as plt
%matplotlib
import seaborn as sns


#그래프 한글띄우기
import pandas as pd
import matplotlib.pyplot as plt
import platform
if platform.system() == 'Darwin': 
        plt.rc('font', family='AppleGothic') 
elif platform.system() == 'Windows': 
        plt.rc('font', family='Malgun Gothic') 
elif platform.system() == 'Linux': 
        plt.rc('font', family='Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False 

plt.figure(figsize = (8,6))
sns.barplot(data = df_new_j, x = 'statsTrgtNm', y='centerNm')
plt.xlabel('직업', size = 15)
plt.ylabel('인원(명)', rotation = 360, size = 15)
plt.show()
```
