```python
import pandas as pd
data = pd.read_csv('흡연_음주 여부/smoke_past.csv')
data
```

```python
data.loc[data['statsTrgtNm'] == 'Y']
```

```python
df = data[['ptSexCd', 'statsTrgtNm']]
df
```

```python
from sklearn.metrics import matthews_corrcoef

df['gender']=df['ptSexCd'].map({'M':1, 'F':2})
df['yub']=df['statsTrgtNm'].map({'Y':1, 'N':0})
df
```

```python
# 남성 과거 흡연율과 폐암의 상관관계 
from sklearn.metrics import matthews_corrcoef

y_true = df.gender
y_pred = df.yub
matthews_corrcoef(y_true, y_pred)
```
