# 문제 1

영화이름 input으로 받고  runtime을 return하는 함수

````python
import requests
import json
import ast
search_name = str(input("제목을 입력하세요 : "))

multi_search_api = "https://api.themoviedb.org/3/search/multi?api_key=eac136b19bd53abf1ad95a1444efa985&language=ko&query=%s" %search_name
r = requests.get(multi_search_api)
data = json.loads(r.text)
data ['results'] = data['results'][0]
movie_id = int(data ['results']['id'])

movie_api = "https://api.themoviedb.org/3/movie/%d?api_key=eac136b19bd53abf1ad95a1444efa985&language=ko" %movie_id
r = requests.get(movie_api)
data2 = json.loads(r.text)
print("runtime :",data2['runtime'])
````

# 문제3

input(name, season, episode) -> output(runtime)

````python
import requests
import json
import ast

search_name = str(input("tv 제목을 입력하세요 : "))

multi_search_api = "https://api.themoviedb.org/3/search/multi?api_key=eac136b19bd53abf1ad95a1444efa985&language=ko&query=%s" %search_name
r = requests.get(multi_search_api)
data = json.loads(r.text)
result = data['results'][0]
tv_id = int(result['id'])

tv_api = "https://api.themoviedb.org/3/tv/%d?api_key=eac136b19bd53abf1ad95a1444efa985" %tv_id

r2 = requests.get(tv_api)
data2 = json.loads(r2.text)

run_time = sum(data2['episode_run_time']) / len(data2['episode_run_time'])
print("runtime :",run_time)
````
