```python
gu_city_name = list(info_df['시군구명'].unique())
gu_city_name
```

```python

df_list=[]

for gu_city in gu_city_name:
    filter = info_df['시군구명'] == gu_city
    print(gu_city+'\n')
    temp = info_df[filter].groupby('읍면동명')['결제건수'].sum().to_frame()
    df_list.append(temp)
    # 파일저장 코드
    print(temp)
    print('\n\n\n')
    
```
