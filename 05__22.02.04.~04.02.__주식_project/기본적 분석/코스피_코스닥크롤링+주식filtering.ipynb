```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
# from openpyxl import Workbook

list_1 = []
jongmok = []
for i in range(1,36):
    url = "https://finance.naver.com/sise/sise_market_sum.naver?&page={}".format(i)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    Kospi_200 = soup.find("table", attrs={"class": "type_2"}).find("tbody").find_all("a", attrs={"class": "tltle"})

    for index, stock in enumerate(Kospi_200):
        jongmok.append(stock.text)
        link = "https://finance.naver.com/"+stock["href"]     # a tag 내에서 "href" 속성값을 가져온다.
        sub_res = requests.get(link)                                 # 링크를 통해 우리가 원하는 기업별 데이터 페이지 데이터 크롤링
        sub_soup = BeautifulSoup(sub_res.text, 'lxml')
    
        ParamList = ['매출액', '영업이익', '당기순이익','ROE(지배주주)','EPS(원)','BPS(원)']
        
        Stock_contents = []
        for idx, pText in enumerate(ParamList):
            try:
                param = " ".join(sub_soup.find('strong', text=pText).parent['class'])
                sub_tbody = sub_soup.find("table", attrs={"class": "tb_type1 tb_num tb_type1_ifrs"}).find("tbody")
                sub_title = sub_tbody.find("th", attrs={"class": param}).get_text().strip()
    
                #param 에 매핑되는 row 검색 => 상위 이동 => 해당 row의 모든 td 컬럼 가져오기
                dataOfParam = sub_tbody.find("th", attrs = {"class":param}).parent.find_all("td")

                for i in dataOfParam:
                    Stock_contents.append(i.get_text().strip())
                    #print(sub_title, " : ",value_param)
            except:
                continue
        
        list_1.append(Stock_contents)
      
```

```python
prices_2021 = pd.read_excel('../gdrive/MyDrive/stock/2021_12_30_prices.xlsx')[['종목코드','종목명','종가']]
prices_2021.columns = ['종목코드','종목명','종가_2021']
```

```python
per_lower = join_t3.loc[join_t3['업종PER']>join_t3['PER']]
per_lower = pd.merge(per_lower,prices_2021,how='left',on='종목코드')[['종목명_x','업종PER','PER','종가','종가_2021']]
per_lower['증가율'] = np.round(((per_lower['종가_2021']-per_lower['종가'])/per_lower['종가'])*100,2)
per_lower = per_lower.sort_values('PER')
```

```python
def x10(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
per_lower['상승여부'] = per_lower['증가율'].apply(x10)
```

```python
per_lower['증가율'].sum()/len(per_lower) #  업종 PER 보다 낮았던 종목들의 3년 평균 주식증가율
```

```python
per_higher = join_t3.loc[join_t3['업종PER']<join_t3['PER']]
per_higher = pd.merge(per_higher,prices_2021,how='left',on='종목코드')[['종목명_x','업종PER','PER','종가','종가_2021']]
per_higher['증가율'] = np.round(((per_higher['종가_2021']-per_higher['종가'])/per_higher['종가'])*100,2)
per_higher = per_higher.sort_values('PER')
per_higher['상승여부'] = per_higher['증가율'].apply(x10)
```

```python
# 3년간 주가 평균 증가율
per_higher['증가율'].sum()/len(per_higher) #  업종 PER 보다 높았던 종목들의 3년 평균 주식증가율
```
